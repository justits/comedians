```python
import pandas as pd
import os
import re

BASE_DIR = '../../'
```

```python
info_path = os.path.join(BASE_DIR, 'data', 'raw_data', 'info_overclocking.csv')
info = pd.read_csv(info_path)
info.head()
```

```python
info.info()
```

```python
info['published_at'] = pd.to_datetime(info.published_at)
```

```python
def num_of_video(name):
    num_descr = name.split(' ')[1]
    num_search = re.search("#\d{1,}", name)
    return int(num_search.group()[1:]) if num_search else 0

info['number'] = info \
    .apply(lambda x: num_of_video(x.title),
           axis=1) \
    .astype('int')
```
